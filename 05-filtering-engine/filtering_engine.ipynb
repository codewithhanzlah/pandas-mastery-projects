## **Import**

```python
import pandas as pd
```

## **Load Dataset**

```python
df = pd.read_csv("data/titanic.csv")
df.head()
```

## **Simple Filtering**

```python
df[df['age'] > 30]
```

## **Multi-Conditional Filtering**

```python
df[(df['age'] > 30) & (df['sex'] == 'male')]
```

```python
df[(df['age'] < 25) | (df['survived'] == 1)]
```

## **Boolean Masking**

```python
mask = df['fare'] > 50
df[mask]
```

## **Query-Based Filtering**

```python
df.query("age > 30 and sex == 'female'")
```

## **Column-Based Filtering**

```python
df.loc[df['pclass'] == 1, ['age', 'fare', 'sex']]
```

## **Complex Filtering**

```python
df[
    (df['pclass'] == 1) &
    (df['fare'] > 100) &
    (df['survived'] == 1)
]
```

## **Using isin()**

```python
df[df['embarked'].isin(['C', 'S'])]
```

## **Save Filtered Output**

```python
filtered_df = df[(df['age'] > 30) & (df['survived'] == 1)]
filtered_df.to_csv("data/filtered_output.csv", index=False)
```
