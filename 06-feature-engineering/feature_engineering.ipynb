## **Import**

```python
import pandas as pd
```

## **Load Dataset**

```python
df = pd.read_csv("data/titanic.csv")
df.head()
```

## **Create New Features**

```python
df['family_size'] = df['sibsp'] + df['parch'] + 1
```

```python
df['is_alone'] = df['family_size'].apply(lambda x: 1 if x == 1 else 0)
```

## **Feature Transformation**

```python
df['fare_per_person'] = df['fare'] / df['family_size']
```

## **Binning**

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[0, 12, 18, 35, 60, 100],
    labels=['child', 'teen', 'young_adult', 'adult', 'senior']
)
```

## **Encode New Feature**

```python
df['age_group_encoded'] = df['age_group'].cat.codes
```

## **Apply Lambda**

```python
df['high_fare'] = df['fare'].apply(lambda x: 1 if x > 50 else 0)
```

## **Check New Feaures**

```python
df[['family_size', 'is_alone', 'fare_per_person', 'age_group', 'high_fare']].head()
```

## **Save Dataset**

```python
df.to_csv("data/titanic_featured.csv", index=False)
```
