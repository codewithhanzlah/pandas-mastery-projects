## **Import**

```python
import pandas as pd
import numpy as np
```

## **Create Dataset**

```python
np.random.seed(42)

data = {
    "student": [f"Student_{i}" for i in range(1, 101)],
    "class": np.random.choice(["A", "B", "C"], 100),
    "subject": np.random.choice(["Math", "Science", "English"], 100),
    "marks": np.random.randint(30, 100, 100)
}

df = pd.DataFrame(data)
df.head()
```

## **Save Dataset**

```python
df.to_csv("data/student_data.csv", index=False)
```

## **Load Dataset**

```python
df = pd.read_csv("data/student_data.csv")
```

## **Average Marks per Subject**

```python
avg_marks = df.groupby("subject")["marks"].mean().reset_index()
avg_marks
```

## **Average Marks per Class**

```python
class_avg = df.groupby("class")["marks"].mean().reset_index()
class_avg
```

## **Topper(Highest Marks)**

```python
topper = df.sort_values(by="marks", ascending=False).head(1)
topper
```

## **Topper per Subject**

```python
subject_topper = df.loc[df.groupby("subject")["marks"].idxmax()]
subject_topper
```

## **Pass/Fail Column**

```python
df["result"] = df["marks"].apply(lambda x: "Pass" if x >= 40 else "Fail")
```

## **Pass/Fail Ratio**

```python
pass_fail_ratio = df["result"].value_counts()
pass_fail_ratio
```

## **Pass Percentage per Class**

```python
pass_percentage = (
    df[df["result"] == "Pass"]
    .groupby("class")
    .size() / df.groupby("class").size()
) * 100

pass_percentage.reset_index(name="pass_percentage")
```

## **Save Final Dataset**

```python
df.to_csv("data/student_processed.csv", index=False)
```
