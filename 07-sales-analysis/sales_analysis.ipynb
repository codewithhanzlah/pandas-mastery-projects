## **Import**

```python
import pandas as pd
import numpy as np
```

## **Create Dataset**

```python
np.random.seed(42)

data = {
    "city": np.random.choice(["Lahore", "Karachi", "Islamabad"], 100),
    "product": np.random.choice(["Laptop", "Phone", "Tablet"], 100),
    "sales": np.random.randint(100, 1000, 100)
}

df = pd.DataFrame(data)

df.head()
```

## **Save Dataset**

```python
df.to_csv("data/sales_data.csv", index=False)
```

## **Load Dataset**

```python
df = pd.read_csv("data/sales_data.csv")
```

## **Sales by City**

```python
city_sales = df.groupby("city")["sales"].sum().reset_index()
city_sales
```

## **Sales by Product**

```python
product_sales = df.groupby("product")["sales"].sum().reset_index()
product_sales
```

## **City + Product Analysis**

```python
city_product_sales = df.groupby(["city", "product"])["sales"].sum().reset_index()
city_product_sales
```

## **Average Sales**

```python
avg_sales = df.groupby("city")["sales"].mean().reset_index()
avg_sales
```

## **Best Performing City**

```python
best_city = city_sales.sort_values(by="sales", ascending=False).head(1)
best_city
```

## **Worst Performing City**

```python
worst_product = product_sales.sort_values(by="sales").head(1)
worst_product
```
