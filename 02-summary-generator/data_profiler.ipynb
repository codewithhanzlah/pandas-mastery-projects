## **Import**

```python
import pandas as pd
```

## **Load Dataset**

```python
file_path = "data/titanic.csv"
df = pd.read_csv(file_path)

df.head()
```

## **Basic Summary Stats**

```python
df.describe()
```

```python
df.describe(include='object')
```

## **Mean,Median,Std**

```python
summary_stats = df.select_dtypes(include=['number']).agg(['mean', 'median', 'std'])
summary_stats
```

## **Unique Values Per Column**

```python
df.nunique()
```

## **Value Counts**

```python
df['Sex'].value_counts()
```

```python
df['Pclass'].value_counts()
```

## **Missing Values Report**

```python
missing = df.isnull().sum()
missing
```

```python
missing_percent = (missing / len(df)) * 100
missing_percent
```

## **Clean Summary Table**

```python
summary_report = pd.DataFrame({
    "Column": df.columns,
    "Data Type": df.dtypes.values,
    "Missing Values": df.isnull().sum().values,
    "Missing %": ((df.isnull().sum() / len(df)) * 100).values,
    "Unique Values": df.nunique().values
})

summary_report
```

```python
summary_report.sort_values(by="Missing %", ascending=False)
```
