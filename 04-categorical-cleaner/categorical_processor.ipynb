## **Import**

```python
import pandas as pd
import seaborn as sns
```

## **Load Dataset**

```python
df = sns.load_dataset("titanic")
df.head()
```

## **Select Categorical Columns**

```python
cat_cols = df.select_dtypes(include=['object', 'category']).columns
cat_cols
```

## **Text Cleaning**

```python
for col in cat_cols:
    df[col] = df[col].astype(str).str.lower().str.strip()
```

## **Value Standardization (Example)**

```python
# Standardizing 'sex' column
df['sex'] = df['sex'].replace({
    'male': 'male',
    'female': 'female'
})
```

## **Label Encoding (Binary Columns)**

```python
df['sex'] = df['sex'].map({'male': 1, 'female': 0})
```

## **One-Hot Encoding**

```python
df_encoded = pd.get_dummies(df, columns=['embarked', 'class'], drop_first=True)
df_encoded.head()
```

## **Check Result**

```python
df_encoded.info()
```

## **Save Processed Data**

```python
df_encoded.to_csv("data/titanic_processed.csv", index=False)
```
